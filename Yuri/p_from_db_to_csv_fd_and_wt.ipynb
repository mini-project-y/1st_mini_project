# DB저장 데이터를 CSV로 변환

### db에서 불러오기

```python
import pandas as pd
from pandas.api.types import CategoricalDtype
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pymysql
# from sqlalchemy import create_engine
```

```python
# 1. 데이터베이스 연결 -> 연결 객체 반환
conn = pymysql.connect(host="localhost", port=3306, db="mp",
                       user="root", passwd="1234", cursorclass=pymysql.cursors.DictCursor)

cursor = conn.cursor()
```

```python
# 3. SQL 작성 + 명령 실행 ( 조회 명령 ) - 미세먼지
sql = """select * from fine_dust"""
cursor.execute(sql)
result_fd = cursor.fetchall()
result_fd
```

```python
fine_dust_df = pd.DataFrame(result_fd)
fine_dust_df.head()
```

```python
# 3. SQL 작성 + 명령 실행 ( 조회 명령 ) - 날씨
sql = """select * from weather_data"""
cursor.execute(sql)
result_wt = cursor.fetchall()
result_wt
```

```python
weather_df = pd.DataFrame(result_wt)
weather_df.head()
```

```python
conn.close()
cursor.close()
```

### DB에서 불러온 데이터를 csv파일로 저장

```python
fine_dust_df.to_csv('./data/fine_dust.csv')
weather_df.to_csv('./data/weather_data.csv')
```
