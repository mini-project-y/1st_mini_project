```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pymysql
# from sqlalchemy import create_engine
```

```python
# 1. 데이터베이스 연결 -> 연결 객체 반환
conn = pymysql.connect(host="localhost", port=3306, db="mp",
                       user="root", passwd="1234", cursorclass=pymysql.cursors.DictCursor)

cursor = conn.cursor()
```

```python
# 3. SQL 작성 + 명령 실행 ( 조회 명령 )
sql = """select * from weather_data"""
cursor.execute(sql)
result = cursor.fetchall()
result
```

```python
weather_df = pd.DataFrame(result)
weather_df.head()
```

```python
conn.close()
cursor.close()
```

```python
weather_df.to_csv('./data/weather_data.csv')
```
