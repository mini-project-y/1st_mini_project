```python
# 전체 df를 컬럼별로 자르기
w_df = df.loc[:,['지점명', '일시', '평균기온(°C)', '최저기온(°C)', '최고기온(°C)', '일강수량(mm)', 
                '평균 풍속(m/s)', '합계 일사량(MJ/m2)', '평균 전운량(1/10)']]
w_df.head(1)
```

```python
w_df = w_df.fillna(0)
print(w_df.info())
# 미세먼지 정보 필요
w_df.rename(columns={'일시':'측정일시'}, inplace=True)
w_df['측정일시'] = w_df['측정일시'].astype('str')
# w_df['측정일시'] = pd.to_datetime(w_df['측정일시'])
print(w_df.info())
```
