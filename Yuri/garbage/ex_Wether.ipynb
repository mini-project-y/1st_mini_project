```python
df2 = df.loc[:,['지점명', '일시', '평균기온(°C)', '최저기온(°C)', '최고기온(°C)', 
                '일강수량(mm)', '평균 상대습도(%)', '합계 일사량(MJ/m2)', '평균 풍속(m/s)']]
df2 = df2.fillna(0)
print(df2.info())
print(df2)
# 미세먼지 정보 필요
df2.columns = ['Area', 'Date', 'Ave_Tem', 'Min_Tem', 'Max_Tem', 'Precipitation',
               'Humidity', 'Insolation', 'Air_velocity']
df2.columns
```

```python
good_df = df2[(df2['Precipitation'] == 0) & (df2['Min_Tem'] >= 15) 
                             & (df2['Max_Tem'] <= 28) & (df2['Humidity'].map(lambda v: v >= 40 and v <= 65)) 
                             & (df2['Air_velocity'] < 4)]
good_df
```

```python
good_df.to_csv("./wether.csv")
```
