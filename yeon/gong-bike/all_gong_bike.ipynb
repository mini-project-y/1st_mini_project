```python
# CSV 파일 읽기
df_bikes = pd.read_csv('merged_bike_data.csv')
df = pd.read_csv('combined_data.csv', encoding='utf-8')

# '일시' 또는 'STAT_DATA'와 같이 공통된 열을 기준으로 병합하기
merged_df = pd.merge(df_bikes, df, left_on='STAT_DATA', right_on='일시', how='inner')  # inner, left, right, outer 옵션 선택 가능

# 결과 확인
merged_df.head(20)
```

```python
a = merged_df[['STAT_DATA', 'RENT_CNT']]
```

```python
a = a.groupby('STAT_DATA').sum()
```

```python
a
```

```python
bike = a.reset_index()
```

```python
bike.to_csv('bike_data.csv', index=False, encoding='utf-8')
print(bike)
```
