```python
# CSV 파일 읽기
df_bikes = pd.read_csv('../bike-data/merged_bike_data.csv')
df = pd.read_csv('combined_data.csv', encoding='utf-8')

# '일시' 또는 'STAT_DATA'와 같이 공통된 열을 기준으로 병합하기
merged_df = pd.merge(df_bikes, df, left_on='STAT_DATA', right_on='일시', how='left')  # inner, left, right, outer 옵션 선택 가능

merged_df.to_csv('D:/yeon/project/1st_miniproject/yeon/merged_dfdfdf.csv')

# 결과 확인
merged_df.head(20)
```

```python
# 대여일자 별 대여건수
merged_sum = merged_df.groupby('대여일자')['대여건수'].sum().reset_index()
```

```python
merged_sum
```

```python
# 2022년, 2023년, 2024년 월별 대여 건수 비교

# 연도와 월 추출
merged_sum['대여년도'] = merged_sum['대여일자'].dt.year
merged_sum['대여월'] = merged_sum['대여일자'].dt.month
```

```python
# 연도별 월별 대여건수 합계 계산
monthly_rentals = merged_sum.groupby(['대여년도', '대여월'])['대여건수'].sum().reset_index()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'  # Windows의 경우
plt.rcParams['axes.unicode_minus'] = False

# 시각화
plt.figure(figsize=(12, 8))
sns.lineplot(x='대여월', y='대여건수', hue='대여년도', data=monthly_rentals, marker='o')
plt.title('2022년, 2023년, 2024년 월별 대여 건수 비교')
plt.xlabel('월')
plt.ylabel('대여건수')
plt.xticks(range(1, 13))  # 월을 1월부터 12월까지 표시
plt.legend(title='대여년도')
plt.grid()
plt.show()
```

```python
# 대여일자에서 년도와 월 추출
merged_df['년도'] = merged_df['대여일자'].dt.year
merged_df['월'] = merged_df['대여일자'].dt.month
```

```python
# 대여일자에서 요일 추출 (0=월요일, 6=일요일)
merged_df['요일'] = merged_df['대여일자'].dt.dayofweek

# 평일(0~4)과 주말(5,6) 구분: 평일은 0, 주말은 1
merged_df['평일주말'] = merged_df['요일'].apply(lambda x: 1 if x >= 5 else 0)
```

```python
# 년도별, 월별, 평일/주말별 대여 건수 합계 계산
monthly_weekday_weekend_rentals = merged_df.groupby(['년도', '월', '평일주말'])['대여건수'].sum().reset_index()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 폰트 설정 (필요 시)
plt.rcParams['font.family'] = 'Malgun Gothic'  # Windows에서 사용 가능
plt.rcParams['axes.unicode_minus'] = False

# 시각화
plt.figure(figsize=(14, 8))
sns.barplot(x='월', y='대여건수', hue='평일주말', data=monthly_weekday_weekend_rentals, palette='viridis', 
            ci=None, hue_order=[0, 1])

# 제목과 레이블 설정
plt.title(' 월별 평일과 주말 대여 건수 비교')
plt.xlabel('월')
plt.ylabel('대여 건수')
plt.xticks(range(12), ['1월', '2월', '3월', '4월', '5월', '6월', '7월', '8월', '9월', '10월', '11월', '12월'])
plt.legend(title='평일/주말', labels=['평일', '주말'])
plt.show()
```

# 요일별 대여건수 분석

```python
import pandas as pd

# 대여일자로부터 요일 추출
merged_df['요일'] = pd.to_datetime(merged_df['대여일자']).dt.day_name()

# 평일/주말 구분 컬럼 (주말을 'Saturday', 'Sunday'로 구분)
merged_df['주말_평일'] = merged_df['요일'].apply(lambda x: '주말' if x in ['Saturday', 'Sunday'] else '평일')

# 평일과 주말의 요일별 대여건수 합계 계산
rental_by_day_type = merged_df.groupby(['주말_평일', '요일'])['대여건수'].sum().reset_index()

# 결과 확인
rental_by_day_type
```

```python
# 요일별 대여건수 시각화하기
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 글꼴 설정
plt.rc('font', family='Malgun Gothic')

# 요일 순서 설정
day_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

# 요일별 평일/주말 대여건수 시각화
plt.figure(figsize=(10, 6))
sns.barplot(data=rental_by_day_type, x='요일', y='대여건수', hue='주말_평일', order=day_order, palette=['orange', 'blue'])

# 그래프 제목 및 라벨 설정
plt.xlabel('요일')
plt.ylabel('대여건수')
plt.title('평일과 주말의 요일별 대여건수 비교')
plt.legend(title='주말/평일')

plt.show()
```

```python
#주말/공휴일별 대여 건수 변화

# 주말/공휴일별 대여 건수의 월별 변화
merged_df['월'] = merged_df['대여일자'].dt.month
weekend_holiday_monthly_rentals = merged_df.groupby(['월', '주말_공휴일'])['대여건수'].sum().reset_index()

# 시각화
plt.figure(figsize=(14, 8))
sns.lineplot(x='월', y='대여건수', hue='주말_공휴일', data=weekend_holiday_monthly_rentals, marker='o')
plt.title('주말/공휴일별 월별 대여 건수 변화')
plt.xlabel('월')
plt.ylabel('대여 건수')
plt.xticks(range(1, 13), ['1월', '2월', '3월', '4월', '5월', '6월', '7월', '8월', '9월', '10월', '11월', '12월'])
plt.show()
```

```python
import pandas as pd

# 예제: 대여소별 대여건수 합계 계산
rental_counts_by_station = merged_df.groupby('대여소명')['대여건수'].sum().reset_index()

# 대여 건수가 많은 상위 10개 대여소
top_10_stations = rental_counts_by_station.nlargest(10, '대여건수')
print("대여 건수가 가장 많은 상위 10개 대여소")
(top_10_stations)
```

```python
# 대여 건수가 적은 하위 10개 대여소
bottom_10_stations = rental_counts_by_station.nsmallest(10, '대여건수')
print("\n대여 건수가 가장 적은 하위 10개 대여소")
(bottom_10_stations)
```

```python
# 대여소별 대여 수요의 차이 파악


import matplotlib.pyplot as plt
import seaborn as sns

# 한글 폰트 설정 (필요 시)
plt.rcParams['font.family'] = 'Malgun Gothic'  # Windows의 경우
plt.rcParams['axes.unicode_minus'] = False     # 마이너스 표시 깨짐 방지

# Figure와 서브플롯 설정
fig, axes = plt.subplots(1, 2, figsize=(18, 8), gridspec_kw={'width_ratios': [1, 1]})

# 상위 10개 대여소 막대 그래프
sns.barplot(x='대여건수', y='대여소명', data=top_10_stations, palette='viridis', ax=axes[0])
axes[0].set_title('대여 건수가 가장 많은 상위 10개 대여소')
axes[0].set_xlabel('대여 건수')
axes[0].set_ylabel('대여소명')

# 하위 10개 대여소 막대 그래프
sns.barplot(x='대여건수', y='대여소명', data=bottom_10_stations, palette='magma', ax=axes[1])
axes[1].set_title('대여 건수가 가장 적은 하위 10개 대여소')
axes[1].set_xlabel('대여 건수')
axes[1].set_ylabel('')

# 그래프 간격 및 보기 좋게 정리
plt.tight_layout()
plt.show()
```
