```python
import pandas as pd
df_bus = pd.read_csv("D:/WORKSPACE/project/1st_miniproject/Mook/서울시_행정동별_버스_총_승차_승객수_정보.csv", encoding='cp949')
df_train = pd.read_csv("D:/WORKSPACE/project/1st_miniproject/Mook/서울시_행정동별_지하철_총_승차_승객수_정보.csv", encoding='cp949')
```

```python
df_train = df_train.fillna(0)
```

```python
bus_sum = df_bus.groupby('기준_날짜').sum().reset_index()
train_sum = df_train.groupby('기준_날짜').sum().reset_index()
```

```python
bus_sum
```

```python
train_sum
```

```python
bus_sum['total_passenger'] = bus_sum.iloc[:, 3:29].sum(axis=1)
bus_sum['commute_passenger'] = bus_sum['버스_승객_수_07시'] + bus_sum['버스_승객_수_08시'] + bus_sum['버스_승객_수_09시']\
                            + bus_sum['버스_승객_수_18시'] + bus_sum['버스_승객_수_19시'] + bus_sum['버스_승객_수_20시']

train_sum['total_passenger'] = train_sum.iloc[:, 3:29].sum(axis=1)
train_sum['commute_passenger'] = train_sum['지하철_승객_수_07시'] + train_sum['지하철_승객_수_08시'] + train_sum['지하철_승객_수_09시']\
                            + train_sum['지하철_승객_수_18시'] + train_sum['지하철_승객_수_19시'] + train_sum['지하철_승객_수_20시']
```

```python
bus_sum = bus_sum.drop(['버스_승객_수_07시', '버스_승객_수_08시', '버스_승객_수_09시', \
                      '버스_승객_수_18시', '버스_승객_수_19시', '버스_승객_수_20시'], axis=1)

train_sum = train_sum.drop(['지하철_승객_수_07시', '지하철_승객_수_08시', '지하철_승객_수_09시', \
                      '지하철_승객_수_18시', '지하철_승객_수_19시', '지하철_승객_수_20시'], axis=1)
```

```python
bus_sum['other_passenger'] = bus_sum.iloc[:, 3:21].sum(axis=1)
train_sum['other_passenger'] = train_sum.iloc[:, 3:21].sum(axis=1)
```

```python
bus_sum = bus_sum[['기준_날짜', 'total_passenger', 'commute_passenger', 'other_passenger']]
train_sum = train_sum[['기준_날짜', 'total_passenger', 'commute_passenger', 'other_passenger']]
```

```python
bus_sum = bus_sum.rename(columns={'기준_날짜': 'date'})
train_sum = train_sum.rename(columns={'기준_날짜': 'date'})
```

```python
bus_merged.set_index('date', inplace=True)
bus_monthly = bus_merged.resample('ME').sum(numeric_only=True)

train_merged.set_index('date', inplace=True)
train_monthly = train_merged.resample('ME').sum(numeric_only=True)
```

```python
bus_monthly2 = bus_monthly.reset_index()
train_monthly2 = train_monthly.reset_index()
```

```python
bus_monthly2["date_month"] = bus_monthly2['date'].map(lambda v: str(v)[:7])
train_monthly2["date_month"] = train_monthly2['date'].map(lambda v: str(v)[:7])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False

plt.figure(figsize=(14, 7))
sns.lineplot(data=bus_monthly2, x='date_month', y='total_passenger', marker='o', label = 'bus')
sns.lineplot(data=train_monthly2, x='date_month', y='total_passenger', marker='o', label = 'train')
plt.title('Monthly Passenger Count by Transportation Mode')
plt.xlabel('Month')
plt.ylabel('Total Passenger Count')
plt.grid()
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
bus_avg = bus_merged.groupby('holiday_or_weekend')['total_passenger'].mean().reset_index()
bus_avg['holiday_or_weekend'] = bus_avg['holiday_or_weekend'].map({0: 'Weekday', 1: 'holiday_or_weekend'})

train_avg = train_merged.groupby('holiday_or_weekend')['total_passenger'].mean().reset_index()
train_avg['holiday_or_weekend'] = train_avg['holiday_or_weekend'].map({0: 'Weekday', 1: 'holiday_or_weekend'})
```

```python
bus_merged = bus_merged.reset_index()
train_merged = train_merged.reset_index()
```

```python
from pprint import pprint

bus_passengers_description = {}

for holiday_status in [0, 1]:
    bus_passengers_description[holiday_status] = bus_merged[
        (bus_merged['holiday_or_weekend'] == holiday_status)
    ]['total_passenger'].describe()

pprint(bus_passengers_description)
```

```python
from pprint import pprint

train_passengers_description = {}

for holiday_status in [0, 1]:
    train_passengers_description[holiday_status] = train_merged[
        (train_merged['holiday_or_weekend'] == holiday_status)
    ]['total_passenger'].describe()

pprint(train_passengers_description)
```

```python
statistics_bus = pd.DataFrame(bus_passengers_description).T
statistics_bus = statistics_bus[['mean', 'max', 'min']]

# 버스 시각화
statistics_bus.index = ['평일 (0)', '주말&공휴일 (1)']
statistics_bus.plot(kind='bar', figsize=(10, 6))
plt.title('버스 승객 수 통계')
plt.xticks(rotation=0)
plt.ylabel('승객 수')
plt.grid(axis='y')
plt.show()
```

```python
# 통계 값 추출
statistics_train = pd.DataFrame(train_passengers_description).T
statistics_train = statistics_train[['mean', 'max', 'min']]

# 지하철 시각화
statistics_train.index = ['평일 (0)', '주말&공휴일 (1)']
statistics_train.plot(kind='bar', figsize=(10, 6))
plt.title('지하철 승객 수 통계')
plt.xticks(rotation=0)
plt.ylabel('승객 수')
plt.grid(axis='y')
plt.show()
```
